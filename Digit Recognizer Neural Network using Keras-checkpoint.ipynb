### Why Keras on top of TensorFlow?
Keras is a layer on top of TensorFlow that makes things a lot easier. Not only is it easier to use, it's easier to tune.

First I'll import all the stuff I need, which will initialize Keras as a side effect:

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop, SGD, Adagrad, Adadelta, Adam #Imported new optimizers after RMSprop
```

### Downloading the data

```python
import pandas as pd
input_file = "C:/Users/jangn/CODE/Sundog_DataScience/DataScience/DataScience-Python3/data_sets/DigitRecognizer_KAGGLE/train.csv"

df_train=pd.read_csv(input_file)
df_train.shape # (42000, 785)
df_train.head()
```

```python
import pandas as pd
input_file = "C:/Users/jangn/CODE/Sundog_DataScience/DataScience/DataScience-Python3/data_sets/DigitRecognizer_KAGGLE/test.csv"

df_test=pd.read_csv(input_file)
df_test.shape # gives (28000, 784)
df_test.head() #gives a dataframe
```

### Making a Train-Test -split
The original test.csv -file does not contain any label data; it cannot be used in testing the accuracy of the model. I need therefore  to use the original train.csv -file for creating a Train- and Test dataset. Using a standard 80/20 -split the sizes of the files will be as follows<br><br>
Train: 0.80 x 42000 = 33600<br>
Test: 0.20 x 42000 = 8400

```python
from sklearn.model_selection import train_test_split # the needed split-function imported from scikit-learn

train_set, test_set = train_test_split(df_train, test_size=0.20, random_state=42)

X_train_set = train_set.drop(['label'], axis=1) #Dropping 'label', the predicted variable 
y_train_set = train_set['label'] # keeping 'label', the predicted variable 

X_test_set = test_set.drop(['label'], axis=1)
y_test_set = test_set['label']

# Double checking with the shape-function that the results are as expected:

# X_train_set.shape = (33600, 784)
# y_train_set.shape = (33600,)

# X_test_set.shape = (8400, 784)
# y_test_set.shape =n (8400,)
```

### Converting the new Train & Test dataframes created from *train.csv* -file into numpy arrays 

```python
df_train_label_array = y_train_set.as_matrix() #creates a numpy array of the df
df_train_image_array = X_train_set.as_matrix() #creates a numpy array of the df

df_test_image_array = X_test_set.as_matrix()
df_test_label_array = y_test_set.as_matrix()

# Again, doubel checking with the shape-function that the results are as expected also for the numpy arrays:
#df_train_label_array.shape # (33600,)
#df_train_image_array.shape # (33600, 784)

#df_test_image_array.shape # (8400, 784)
#df_test_label_array.shape # (8400,)
```

Next, the data needs to be converted into the format Keras / TensorFlow expects. The image data is divided by 255 in order to normalize it into 0-1 range, after converting it into floating point values.

```python
train_images = df_train_image_array.reshape(33600, 784) #(33600, 784)
test_images = df_test_image_array.reshape(8400, 784) # (8400, 784)
train_images = train_images.astype('float32')
test_images = test_images.astype('float32')
train_images /= 255
test_images /= 255
#test_images
```

Now we'll convert the 0-9 labels for the Train set into "one-hot" format. No labels for Test set are available; these are what the model will try to predict!

```python
train_labels = keras.utils.to_categorical(df_train_label_array, 10)
test_labels = keras.utils.to_categorical(df_test_label_array, 10)
```

### Sample images
Lets look at a sample of 3 images from Train data set, e.g. the 1111st, the 2222nd and the 3333rd image, with their labels to make sure it looks OK:

```python
import matplotlib.pyplot as plt
%matplotlib inline

def display_sample(num):
    #Print the one-hot array of this sample's label 
    print(train_labels[num]) #df_train_label_array
    #Print the label converted back to a number
    label = train_labels[num].argmax(axis=0) #df_train_label_array
    #Reshape the 768 values to a 28x28 image
    image = train_images[num].reshape([28,28])
    #plt.subplot(4,4,i)
    plt.title('Sample: %d  Label: %d' % (num, label))
    plt.imshow(image, cmap=plt.get_cmap('gray_r'))
    plt.show()
    
display_sample(1111) #the 1111st image in the Training set
display_sample(2222) #the 2222nd image in the Training set
display_sample(3333) #the 3333rd image in the Training set
```

All three images above seem to have the correct label on them. <br>

### Building the actual neural network
Compared to TensorFlow Keras makes life a bit easier. Unlike with TensorFlow, where you'd need to write code for creating placeholders, variables, and defining a bunch of linear algebra for each layer in our neural network we can set up the same layers like this. The input layer of 784 (= 28 x 28 pixels) features feeds into a ReLU layer of 512 nodes, which then goes into 10 nodes with softmax applied. Couldn't be simpler:

```python
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,))) 
model.add(Dropout(0.5))
model.add(Dense(512, activation='relu')) 
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

We can even get a nice description of the resulting model:

```python
model.summary()
```

Setting up our optimizer and loss function is just as simple. We will use the Adam optimizer here. Other choices include Adagrad, SGD, RMSProp, Adam, Adamax, and Nadam. See https://keras.io/optimizers/

```python
model.compile(loss='categorical_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy'])
```

### Adjustment of hyperparameters to speed things up

Training the model is also just one line of code with Keras! Here we'll do 20 epochs with a batch size of 100. Keras makes intensive calculations under the hood and can be rather slow, especially on a CPU, which I'm using. A GPU would do the job faster. <br>
I've limited the model in this version to just 20 epochs. My best result in the Kaggle competition using this model was reached with 50 epochs; 98%<BR>

```python
history = model.fit(train_images, train_labels,
                    batch_size=50, 
                    epochs=20, 
                    verbose=2,
                    validation_data=(test_images, test_labels))
```

### And the accuracy of the model was:

```python
score = model.evaluate(test_images, test_labels, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

In c. 97.7% of the cases the model interpreted the image correctly. 

Below is the code for creating the submission file - DR.csv - which I then uploaded to https://www.kaggle.com/c/digit-recognizer


```python
predictions = model.predict_classes(df_test, verbose=0)

submissions=pd.DataFrame({"ImageId": list(range(1,len(predictions)+1)),
                         "Label": predictions})
submissions.to_csv("DR.csv", index=False, header=True)
```

### Where did the model go wrong?
The Train data in the Kaggle-competition consisted of 42000 images and the Test data of 28000 images. With my model's 97.7% accuracy, this left c. 644 images where the model got it wrong. Just out of cusriosity, let's visualize a few where it got it wrong.

```python
for x in range(1000):
    test_image = test_images[x,:].reshape(1,784)
    predicted_cat = model.predict(test_image).argmax()
    label = test_labels[x].argmax()
    if (predicted_cat != label):
        plt.title('Prediction: %d Label: %d' % (predicted_cat, label))
        plt.imshow(test_image.reshape([28,28]), cmap=plt.get_cmap('gray_r'))
        plt.show()
```

You can see most of the ones it's having trouble with, are images a human would have trouble with as well!
